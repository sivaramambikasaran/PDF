# Gaussian distribution

The probability density function (PDF) of a normal/Gaussian random variable with mean $\mu$ and variance $\sigma^2$ is given by
$$f_X(x) = \dfrac1{\sigma\sqrt{2 \pi}} \exp\bkt{-\dfrac12\bkt{\dfrac{x-\mu}{\sigma}}^2}$$
The PDF of a standard normal random variable $Z$ (whose mean is zero and variance is $1$) is given by
$$f_Z(z) = \dfrac1{\sqrt{2 \pi}} \exp\bkt{-\dfrac{z^2}2}$$
A normal random variable $X$ with mean $\mu$ and variance $\sigma^2$ can be transformed into a standard normal random variable, $Z$, by the transformation
$$Z = \dfrac{X-\mu}{\sigma}$$

```python
import scipy as sp;
import scipy.stats as stats
import scipy.integrate as integrate;

# Writing the Gaussian cdf to markdown
file = open("Gaussian_markdown.txt","w");
X = sp.linspace(0,4,400);
Y = sp.around(stats.norm.cdf(X), decimals=5);
for k in range (0,X.size):
    if (k%10 == 0):
        file.write(str(sp.around(X[k], decimals=1)));
        file.write(' | ');
    file.write(str(Y[k]));
    file.write(' | ');
    if (k%10==9):
        file.write(' \n');
file.close()

# Writing the Gaussian cdf to latex
file = open("Gaussian_latex.txt","w");
for k in range (0,X.size):
    if (k%10 == 0):
        file.write(str(sp.around(X[k], decimals=1)));
        file.write(' & ');
    file.write(str(Y[k]));
    if (k%10==9):
        file.write(' \\\\ ');
        file.write(' \n');
    else:
        file.write(' & ');
file.close()

# Writing the Inverse of the Gaussian cdf to markdown for $p \in [0.5,0.9)$
file = open("Inverse_Gaussian_markdown.txt","w");
X = sp.linspace(0.5,0.89,40);
Y = sp.around(stats.norm.ppf(X), decimals=4);
for k in range (0,X.size):
    if (k%10 == 0):
        file.write(str(sp.around(X[k], decimals=1)));
        file.write(' | ');
    file.write(str(Y[k]));
    file.write(' | ');
    if (k%10==9):
        file.write(' \n');
file.close()

# Writing the Inverse of the Gaussian cdf to latex for $p \in [0.5,0.9)$
file = open("Inverse_Gaussian_latex.txt","w");
for k in range (0,X.size):
    if (k%10 == 0):
        file.write(str(sp.around(X[k], decimals=1)));
        file.write(' & ');
    file.write(str(Y[k]));
    if (k%10==9):
        file.write(' \\\\ ');
        file.write(' \n');
    else:
        file.write(' & ');
file.close()

# Writing the Inverse of the Gaussian cdf to markdown for $p \in [0.9,1)$
file = open("Inverse_Gaussian_1_markdown.txt","w");
X = sp.linspace(0.9,0.999,100);
Y = sp.around(stats.norm.ppf(X), decimals=4);
for k in range (0,X.size):
    if (k%10 == 0):
        file.write(str(sp.around(X[k], decimals=2)));
        file.write(' | ');
    file.write(str(Y[k]));
    file.write(' | ');
    if (k%10==9):
        file.write(' \n');
file.close()

# Writing the Inverse of the Gaussian cdf to latex for $p \in [0.9,1)$
file = open("Inverse_Gaussian_1_latex.txt","w");
for k in range (0,X.size):
    if (k%10 == 0):
        file.write(str(sp.around(X[k], decimals=2)));
        file.write(' & ');
    file.write(str(Y[k]));
    if (k%10==9):
        file.write(' \\\\ ');
        file.write(' \n');
    else:
        file.write(' & ');
file.close()

import matplotlib.pyplot as plt

x = sp.arange(-2.5, 0.5, 0.01)
y = sp.exp(-x*x)
X = sp.arange(-2.5, 2.5, 0.01)
Y = sp.exp(-X*X)
fig, (ax) = plt.subplots(1, 1, sharex=True)

ax.fill_between(x, 0, y);
ax.plot(X,Y,'r-');
ax.set_xlabel('Random variable Z')
ax.set_ylabel('PDF')
ax.set_title('P($Z \leq z$) is the shaded area')
plt.rcParams["figure.figsize"] = [14,6];
plt.show()
```
