# $\chi^2$ distribution

$X$ follows a $\chi^2$ dsitribution with $n$ degrees of freedom, if $X = Z_1^2+Z_2^2+ \cdots + Z_n^2$, where $Z_i$'s are standard normal random variables, i.e., $Z_i \sim \mathcal{N}\bkt{0,1}$.

The probability density function (PDF) of a $\chi^2$ random variable with $n$ degrees of freedom is given by
$$f_X(x) = \dfrac1{2^{k/2} \Gamma(k/2)} x^{k/2-1} e^{-x/2}$$

```python
import scipy as sp;
import scipy.stats as stats
import scipy.integrate as integrate;

# Writing the Gaussian cdf to markdown
file_m = open("chi_square_markdown.txt","w");
file_l   = open("chi_square_latex.txt","w");
X = sp.asarray([0.995,0.99,0.975,0.95,0.9,0.75,0.5,0.25,0.1,0.075,0.05,0.025,0.01,0.005]);
# Y = sp.around(, decimals=5);
for n in sp.linspace(1,50,50):
    Y=sp.around(stats.chi2.ppf(1-X,n),decimals=5);
    file_m.write('| ')
    file_m.write(str(int(n)));
    file_m.write(' | ')
    file_l.write(str(int(n)));
    file_l.write(' & ')
    for k in range(0,X.size):
        file_m.write(str(Y[k]));
        file_m.write(' | ');
        file_l.write(str(Y[k]));
        if (k<X.size-1):
            file_l.write(' & ');
    file_m.write(' \n');
    file_l.write(' \\\\\n');

file_m.close()
file_l.close()

import matplotlib.pyplot as plt

n = 5;
x = sp.arange(4, 20, 0.01)
y = 1/2**(n/2)/sp.special.gamma(n/2)*x**(n/2-1)*sp.exp(-x/2);
X = sp.arange(0, 20, 0.01)
Y = 1/2**(n/2)/sp.special.gamma(n/2)*X**(n/2-1)*sp.exp(-X/2);
fig, (ax) = plt.subplots(1, 1, sharex=True)

ax.fill_between(x, 0, y);
ax.plot(X,Y,'r-');
ax.set_xlabel('Random variable Z')
ax.set_ylabel('PDF of $\chi^2_5$ distribution')
ax.set_title('P($Z \leq z$) is the shaded area')
plt.rcParams["figure.figsize"] = [14,6];
plt.show()
```
