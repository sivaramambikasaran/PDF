# Poisson distribution

The probability mass function (PMF) of a Poisson random variable with mean $\mu$ is given by
$$p_X(n) = e^{-\mu}\dfrac{\mu^n}{n!}$$
The cumulative distribution function (CDF) is hence given by
$$F_X(x) = P\bkt{X \leq x} = \dsum_{n=0}^{\lfloor x \rfloor} e^{-\mu}\dfrac{\mu^n}{n!}$$

```python
import scipy as sp;
import scipy.stats as stats
import scipy.integrate as integrate;

# Writing the Gaussian cdf to markdown
file  = open("Poisson_markdown.txt","w");
MU  = sp.linspace(0.1,1.5,15);
X     = range(1,9);
for k in X:
    file.write('| ');
    file.write(str(k));
    file.write(' | ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size):
        file.write(str(Y[j]));
        file.write(' | ');
    file.write('\n');
file.close()

# Writing the Gaussian cdf to latex
file = open("Poisson_latex.txt","w");
for k in X:
    file.write(str(k));
    file.write(' & ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size-1):
        file.write(str(Y[j]));
        file.write(' & ');
    file.write(str(Y[MU.size-1]));
    file.write('\\\\\n\hline\n');
file.close()

# Writing the Gaussian cdf to markdown
file  = open("Poisson_markdown_1.txt","w");
MU  = sp.linspace(1.6,4,13);
X     = range(1,15);
for k in X:
    file.write('| ');
    file.write(str(k));
    file.write(' | ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size):
        file.write(str(Y[j]));
        file.write(' | ');
    file.write('\n');
file.close()

# Writing the Gaussian cdf to latex
file = open("Poisson_latex_1.txt","w");
for k in X:
    file.write(str(k));
    file.write(' & ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size-1):
        file.write(str(Y[j]));
        file.write(' & ');
    file.write(str(Y[MU.size-1]));
    file.write('\\\\\n\hline\n');
file.close()

# Writing the Gaussian cdf to markdown
file  = open("Poisson_markdown_2.txt","w");
MU  = sp.linspace(4.5,10,12);
X     = range(1,26);
for k in X:
    file.write('| ');
    file.write(str(k));
    file.write(' | ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size):
        file.write(str(Y[j]));
        file.write(' | ');
    file.write('\n');
file.close()

# Writing the Gaussian cdf to latex
file = open("Poisson_latex_2.txt","w");
for k in X:
    file.write(str(k));
    file.write(' & ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size-1):
        file.write(str(Y[j]));
        file.write(' & ');
    file.write(str(Y[MU.size-1]));
    file.write('\\\\\n\hline\n');
file.close()

# Writing the Gaussian cdf to markdown
file  = open("Poisson_markdown_3.txt","w");
MU  = sp.linspace(11,20,10);
X     = range(1,36);
for k in X:
    file.write('| ');
    file.write(str(k));
    file.write(' | ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size):
        file.write(str(Y[j]));
        file.write(' | ');
    file.write('\n');
file.close()

# Writing the Gaussian cdf to latex
file = open("Poisson_latex_3.txt","w");
for k in X:
    file.write(str(k));
    file.write(' & ');
    Y = sp.around(stats.poisson.cdf(k,MU), decimals=5);
    for j in range(0,MU.size-1):
        file.write(str(Y[j]));
        file.write(' & ');
    file.write(str(Y[MU.size-1]));
    file.write('\\\\\n\hline\n');
file.close()
```
